```python
from openai import OpenAI

client = OpenAI(
    api_key="ollama",  
    base_url="http://localhost:11434/v1"
)

SYSTEM_PROMPT = "You're a helpful assistant"

def msg_gpt(prompt):
    messages = [
               {"role": "system", "content": SYSTEM_PROMPT},
               {"role": "user", "content": prompt}
            ]
    
    response = client.chat.completions.create(model="phi", messages=messages)

    return response.choices[0].message.content

print(msg_gpt("Hi my name is Ahmad"))
```

```python
import gradio as gr

def hello(name):
    return "Hello "+ name

input_box = gr.Textbox(lines=8, label="NAME", info="Enter Your name please!")
output_box =  gr.Textbox(lines=8, label="RESPONSE")

demo = gr.Interface(fn= hello, inputs = [input_box], outputs = [output_box])

demo.launch(share=False, inbrowser = True, auth = ("ahmad", "123"))
```
